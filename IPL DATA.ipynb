```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('each_match_records.csv')
```

```python
df.head()
```

# Ball First

```python
 baller_first =df[df['toss_decision']=='field']
```

```python
baller_first['winner'].value_counts()
```

```python
plt.figure(figsize = (20,7))
ax =sns.countplot(x ='winner' , data = baller_first)
ax.set_xticklabels(ax.get_xticklabels() , rotation = 30)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Winning Teams Who Choose Balling')    
plt.show()
```
