# Exploratory Data Analysis

```python
Area=df.nlargest(10,"area")
ax=plt.subplots(figsize=(13,5))
plt.title("Largest Area ")
ax=sns.barplot(x="country",y="area",data=Area)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
Rank=df.nsmallest(10,"rank")
ax=plt.subplots(figsize=(13,5))
plt.title("Ranks ")
ax=sns.barplot(y="country",x="rank",data=Rank)
for bars in ax.containers:
    ax.bar_label(bars)
    
    plt.xlabel("Country")
    plt.ylabel("Rank")
plt.show()
```

```python
growth=df.nlargest(10,"growthRate")
plt.figure(figsize=(10,5))

sns.barplot(y="growthRate",x="country",data=growth)
```

```python
df.info()
```

```python
Density=df.nlargest(10,"density")
plt.figure(figsize=(12,6))
sns.barplot(x="country",y="density",data=Density)
plt.show()
```

```python
countries = ['India', 'China', 'United States', 'United Kingdom', 'Japan']

years = ['pop1980', 'pop2000', 'pop2010', 'pop2023','pop2030','pop2050']  

plt.figure(figsize=(10, 6)) 

for country in countries:
    country_data = df[df['country'] == country]
    plt.plot(years, country_data[years].values.reshape(-1),label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trend for Selected Countries')
plt.legend()
plt.show()
```
