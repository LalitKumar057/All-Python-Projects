```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_html('https://www.investopedia.com/insights/worlds-top-economies/',header = 0)[0]
df.head()
```

```python
df.columns = df.iloc[0]
```

```python
newdf = df.iloc[1:]
```

```python
newdf.head()
```

```python
newdf['Nominal GDP (in trillions)']
```

```python
newdf.head()
```

```python
column_to_change =['Nominal GDP (in trillions)','PPP Adjusted GDP (in trillions)']
```

```python
for column in column_to_change:
    newdf[column]=newdf[column].str.replace('$','').astype(float)
```

```python
newdf['Annual Growth (%)'] = newdf['Annual Growth (%)'].str.replace('%','').astype(float)
newdf['GDP Per Capita'] = newdf['GDP Per Capita'].str.replace('$','')
newdf['GDP Per Capita']=newdf['GDP Per Capita'].str.replace(',','').astype(int)
```

```python
newdf.head()
```

```python
newdf.dtypes
```

# Exploratory Data Analysis

```python
plt.figure(figsize = (10,6))
sns.barplot(x= 'Country' , y = 'Nominal GDP (in trillions)' , data = newdf )
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.barplot(x= 'Country' , y = 'PPP Adjusted GDP (in trillions)' , data = newdf )
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.lineplot(x= 'Country' , y = 'Annual Growth (%)' , data = newdf )
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (10,6))
ax= sns.barplot(y= 'GDP Per Capita' , x = 'Country' , data = newdf)
for bars in ax.containers:
    ax.bar_label(bars)
plt.xticks(rotation = 45)
plt.show()
```
