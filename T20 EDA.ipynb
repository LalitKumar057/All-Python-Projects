```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('t20_match_results.csv')
```

```python
df.head(2)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Winner',data =df)
plt.ylabel('No Of Winings')
plt.title('No Of Winings')
plt.show()
```

**Bat First**

```python
Bat_first=df[df['Margin'].str.contains('runs')]
```

```python
Bat_first.head(2)
```

```python
hist=Bat_first['Margin'].str.replace('runs','').astype(int)
```

```python
sns.histplot(hist)
plt.title('Distribution Of Winnings Runs')
plt.show()
```

**Ball First**

```python
Ball_first=df[df['Margin'].str.contains('wickets')]
```

```python
Ball_first.head(2)
```

```python
ball_hist = Ball_first['Margin'].str.replace('wickets','').astype(int)
```

```python
sns.histplot(ball_hist)
plt.title('Distribution of Winnings wickets')
plt.show()
```

**Ground**

```python
plt.figure(figsize=(10,5))
ax=sns.countplot(x='Ground',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('No Of Matches Played In Ground')
plt.show()
```

```python
ax=sns.countplot(x='Match Date', data= df)
ax.set_xticklabels(ax.get_xticklabels() ,rotation =90)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```
