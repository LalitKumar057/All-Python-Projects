```python
import matplotlib.pyplot  as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Netflix Userbase.csv')
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

# Exploratory Data Analysis(EDA)

**Subcription Type Vs Monthly Revenue**

```python
subs_type =df.groupby(['Subscription Type'] ,as_index = False)['Monthly Revenue'].sum().sort_values(by='Monthly Revenue' ,ascending = False)
```

```python
ax=sns.barplot(x='Subscription Type', y = 'Monthly Revenue' , data = subs_type)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Revenue By Subscription Type')
plt.xlabel('Subscriptions')
plt.ylabel('Renvenue in $')
plt.show()
```

```python
#Subscription Count
sns.countplot(x='Subscription Type',data =df)
plt.title('Count Of Subscription Type')
plt.show()
```

**Revenue By Gender**

```python
gen=df.groupby(['Gender'] ,as_index = False)['Monthly Revenue'].sum().sort_values(by='Monthly Revenue' ,ascending = False)
```

```python
ax =sns.barplot(x='Gender', y = 'Monthly Revenue' , data = gen)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('Revenue By Gender')
plt.show()
```

```python
#Countplot For Male And Female
sns.countplot(x='Gender', data =df,order = ['Female' ,'Male'])
plt.ylabel('No Of Genders')    
plt.show()    
```

```python
#Country Which has More Users
plt.figure(figsize=(15,5))
sns.countplot(x='Country',data=df)
plt.xlabel('No Of Users')
plt.show()
```

**Countries Have Max Users**

```python
#Device Counplot
ax=sns.countplot(x='Device', data =df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(x='Age',data=df)
plt.show()
```
