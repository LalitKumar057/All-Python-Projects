```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("D:\Downloads\creditcard.csv\creditcard.csv")
```

```python
df.head(2)
```

```python
df['Class'].value_counts()
```

**0 for normal_transaction**  ,
**1 for fraud transaction**

```python
legit = df[df['Class']==0]
fraud = df[df['Class']==1]
```

```python
legit.head(2)
```

```python
fraud.head(2)
```

```python
df.groupby(by = 'Class').mean()
```

```python
#coz legit and fraud transaction are not in same distribution , so we create a sample of legit transaction similar to fraud
legit_sample = legit.sample(n=492)
```

```python
#concat the legit_sample and  fraud
newdf  = pd.concat([legit_sample , fraud]  ,axis  =  0)
```

```python
newdf.head(2)
```

```python
newdf['Class'].value_counts()  # that is similar distribution
```

```python
newdf.groupby(by = 'Class').mean()
```

```python
X = newdf.drop('Class' , axis = 1)
Y  = newdf['Class']
```

```python
X.head(2)
```

```python
Y.head(2)
```

```python
X_train  , X_test , Y_train  ,Y_test   = train_test_split(X, Y ,test_size = .20)
```

```python
X_train.shape  , X_test.shape , Y_train.shape  ,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model =  LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction  = model.predict(X_train)
```

```python
training_data_accuracy  = accuracy_score(X_train_prediction , Y_train)
print('training data accuracy is  : '  ,training_data_accuracy)
```

```python
X_test_prediction  = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(X_test_prediction  , Y_test)
print('test data accuracy is  : ' , test_data_accuracy)
```
