```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('final_data.csv')
```

# EDA

```python
#Emp_Count Of Company_Name
emp_count=df.groupby(by='Company_Name',as_index =False)['Total_applicants'].sum().sort_values(by='Total_applicants',ascending=False)
```

```python
sns.barplot(y='Company_Name',x='Total_applicants',data=emp_count)
plt.show()
```

```python
emp_count1=df.groupby(by='Class',as_index =False)['Total_applicants'].sum().sort_values(by='Total_applicants',ascending=False)
```

```python
sns.barplot(x='Class',y='Total_applicants',data=emp_count1)
plt.show()
```

```python
#sns.countplot(x='Designation',data=df)
desg = df['Designation'].value_counts().reset_index()
desg.head(2)
```

```python
desg['count'].astype(int).head(2)
```

```python
desg.dtypes
```

```python
desg1 = desg.nlargest(10,'count')
plt.pie(desg1['count'],labels = desg1['Designation'],autopct = '%0.1f%%')
plt.show()
```

```python
inds = df['Industry'].value_counts().reset_index()
inds['count'].astype(int)
inds1=inds.nlargest(3,'count')
plt.pie(inds1['count'] ,labels = inds1['Industry'],autopct = '%0.1f%%')
plt.show()
```

```python
loc=df.groupby(by='Location',as_index =False)['Total_applicants'].sum().sort_values(by='Total_applicants',ascending=False)
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(y='Location',x='Total_applicants',data=loc)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()    
```

```python
df[['Company_Name','LinkedIn_Followers']].head(2)
```

```python
followers=df.groupby(by='Company_Name',as_index=False)['LinkedIn_Followers'].max().sort_values(by='LinkedIn_Followers',ascending= False)
```

```python
plt.figure(figsize=(10,5))
ax=sns.barplot(y='Company_Name' ,x='LinkedIn_Followers',data =followers)
plt.show()
```

```python
Involve=df.groupby(by='Involvement',as_index =False)['Total_applicants'].sum().sort_values(by='Total_applicants',ascending=False)
```

```python
plt.pie(Involve['Total_applicants'],labels=Involve['Involvement'],autopct='%0.1f%%')
plt.show()
```
