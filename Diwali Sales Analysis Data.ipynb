# Exploratory Data Analysis


**Gender**

```python
df.columns
```

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

***Female have Purchased Many things than Male***

```python
 df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
#Gender vs Amount
sales_gen = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen,errorbar=('ci',0))
plt.show()
```

*From the graph we can canclude that Females have high purchasing power than Males*

**Age**

```python
ax=sns.countplot(x='Age Group',data=df,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
#Age Group vs Amount
sales_gen = df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_gen,)
plt.show()
```

*From the above graph most of the buyers are from the age group 26-35 Females*

**State**

```python
sales_State=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False)
```

```python
ax=plt.subplots(figsize=(25,10))
ax=sns.barplot(x="State",y="Orders",data=sales_State)
plt.ylabel("Orders")
plt.xlabel("State")
plt.show()
```

```python
sales_State=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
ax=plt.subplots(figsize=(25,10))
ax=sns.barplot(x="State",y="Amount",data=sales_State)
plt.ylabel("Amonut")
plt.xlabel("State")
plt.show()
```

*From the graph we can see that most of the orders from UP,Maharastra,Karnatka*

**Marital Status**

```python
ax=sns.countplot(x='Shaadi',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
    
    plt.show()

```

```python
sales_Shaadi=df.groupby(["Shaadi","Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Shaadi",y="Amount",data=sales_Shaadi,hue="Gender")
plt.show()
```

*From the graphs we can see that most of buyers are the married(women), which has high purchasing power*

**Occupation**

```python
ax=plt.subplots(figsize=(20,7))
ax=sns.countplot(x="Occupation",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
sales_Occs=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
ax=plt.subplots(figsize=(20,7))
ax=sns.barplot(x="Occupation",y="Amount",data=sales_Occs)
```

*From the graph we can see that most of the buyers are from the IT sector,Healthcare and aviation*

**Product Category**

```python
ax=plt.subplots(figsize=(25,10))
ax=sns.countplot(x="Product_Category",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Catog =df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
ax=plt.subplots(figsize=(20,8))
ax=sns.barplot(x="Product_Category",y="Amount",data=sales_Catog)

```

*From the graph the most sold product category is Food,Clothing and Apparel,Electronics and gadgets and Footwear and Shoes*

```python
sales_Catog =df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(10)
ax=plt.subplots(figsize=(20,6))
ax=sns.barplot(x="Product_ID",y="Orders",data=sales_Catog)
```

**Conclusion:**

*Married women from UP,Maharastra,Karnatka working in IT,Healthcare and Avaition are more likely to buy products from Food
Clothing and Electronic Category.*
